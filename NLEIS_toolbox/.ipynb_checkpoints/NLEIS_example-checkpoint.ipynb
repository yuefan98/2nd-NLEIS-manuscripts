# 2nd-NLEIS Toolbox (beta version)

This 2nd-NLEIS toolbox follows the design of impedance.py and can be used seamlessly with additional 2nd-NLEIS data.
Here we are sharing an example of 18650 NMC|C cell data. The whole cell model consist of porous cathode and anode with spherical particles. Inductor and resistor are added to the EIS model to adress the measurement artifacts and ohmic resistance. 

To use this 2nd-NLEIS toolbox. Please follow the following instruction to create a virtual environment for it. 

First move to the directory of this toolbox in terminal. Then create an envrionment using the following commend

`conda env create -f environment.yml`

After conda creates this environment, we need to activate it before we can use it by using:

`conda activate impedance`

Weâ€™ve now activated our conda environment and are ready to explore the example notebook

In this NLEIS toolbox. The linear and nonlinear circuit element pair are defined in nonlinear element pairs and the additional 'n' represents nonlinear circuits.  `t` is used to perform subtraction between cathode and anode resposne.

The avaliale pairs are `[TDP,TDPn],[TDS,TDPSn],[TDS,TDSn],[RCD,RCDn] `

The toolbox can be load through the following command

#### Loading the common python package

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore") # ignored the unimportant warnings
```

## Loading the NLEIS example data 

```python
frequencies_30a = np.loadtxt('data/freq_30a.txt')
Z1s_30a = np.loadtxt('data/Z1s_30a.txt').view(complex)
Z2s_30a = np.loadtxt('data/Z2s_30a.txt').view(complex)
```

### 1. Implementation of the linear model in the NLEIS toolbox 

#### data pre-processing

The data can be preprocessed using the `data_processing` function. It will automatically return frequency and impedance pair that truncated the negative Z1.imag, and the data for Z2 to 10Hz. The Z2 truncation can be specified using max_f. The defualt is 10.

```python
from impedance.models.nleis.nleis_fitting import data_processing
```

```python
f = frequencies_30a
Z1 = Z1s_30a[1]
Z2 = Z2s_30a[1]

f1,Z1,f2,Z2,Z2sd = data_processing(f,Z1,Z2)
```

The implementation of NLEIS toolbox can be done with traditional impedance.py syntax. We need to first import CustomCircuit and NLEIS toolbox from impedance.py 

```python
from impedance.models.circuits import CustomCircuit
from impedance.models import nleis
```

The definition of linear circuit for NLEIS should follows first cathode and second anode order if 2nd-NLEIS analysis is expected later. Check out the documentation page for the details of function definition. Other work flow are exactly the same as impedance.py

```python
circ_str_1 = 'L0-R0-TDS0-TDS1'
initial_guess_1 = np.ones(12)*0.01 
circuit_1 = CustomCircuit(circ_str_1,initial_guess=initial_guess_1)
circuit_1.fit(f1,Z1);
circuit_1.plot(f_data=f1, Z_data=Z1, kind='nyquist')
plt.show()
```

### 2. Implementation of the nonlinear models in the NLEIS toolbox

Analysis using nonlinear models in the NLEIS toolbox requires the nonlinear circuit, it is currently done through NLEISCustomCircuit, but it can be fully incorporate into CustomCircuit in the future.

And it has the exact same syntax as adopted in impedance.py

```python
from impedance.models.nleis import NLEISCustomCircuit
```

The function will proceess the data and the default truncation is 10Hz. 

```python
circ_str_2 = 't(TDSn0,TDSn1)'
initial_guess_2= [2.61981964e-02, 5.81684044e-03,.33880313e+00, 2.11669067e-02, 7.75942686e+02,0,0.1, 8.18163535e-03,2.51002220e-03, 3.54916599e-01, 4.62988605e-01, 2.75698184e+04,0,0]
circuit_2 = NLEISCustomCircuit(circ_str_2,initial_guess=initial_guess_2)
circuit_2.fit(f2,Z2);
circuit_2.plot(f_data=f2, Z_data=Z2, kind='nyquist')
plt.show()
```

### 3. Simultaneous analysis of EIS and 2nd-NLEIS using NLEIS toolbox

The simultaneous analysis is enabled by `EISandNLEIS`, which requires input of both linear and nonlinear circuit.
The order of circuit should strictly follows cathode first and anode last for both linear and nonlinear circuit. 

Note: `EISandNLEIS.fit()` will do the data processing automatically, so it is important keep frequencies, Z1, and Z2 have the same length

The default optimization method is max normalization (`opt = 'max'`) with nonlinear least square optimization . negative log likelihood optimization(`opt ='neg'`) is also provided, but it is less stable compared to max normalization.


```python
from impedance.models.nleis import EISandNLEIS
```

```python
f = frequencies_30a
Z1 = Z1s_30a[1]
Z2 = Z2s_30a[1]

f1_data,Z1_data,f2_data,Z2_data,Z2_sd = data_processing(f,Z1,Z2)
```

```python
circ_str_1 = 'L0-R0-TDS0-TDS1'
circ_str_2 = 't(TDSn0,TDSn1)'

initial_guess_3 = [1e-7,1e-3
                   ,5e-3,1e-3,10,1e-2,100,10,0.1
                   ,1e-3,1e-3,1e-3,1e-2,1000,0,0
                   ]
circuit_3 = EISandNLEIS(circ_str_1,circ_str_2,initial_guess=initial_guess_3)
circuit_3.fit(f,Z1,Z2, opt='max');
```

The default style is nyquist, bode plot can also be obtained using `kind = 'bode'`.

`f_data` to be the same length with `Z1_data` which is usually `f1_data`. 

`f1_data`, `Z1_data` and `Z2_data` can be obtained using `data_processing()`

```python
circuit_3.plot(f_data=f1_data, Z1_data =Z1_data, Z2_data= Z2_data, kind='bode')
# plt.legend()
plt.tight_layout()
plt.show()
```

```python
circuit_3.plot(f_data=f1_data, Z1_data =Z1_data, Z2_data= Z2_data, kind='nyquist')
plt.tight_layout()
plt.show()
```

Interactive EIS and NLIES plot can also be obtained with `kind = 'altair'`

```python
EIS_plot,NLEIS_plot = circuit_3.plot(f_data=f1_data, Z1_data =Z1_data, Z2_data= Z2_data, kind='altair')
```

```python
EIS_plot
```

```python
NLEIS_plot
```

`circuit.predict()` can be used to obtain Z1 and Z2 for given frequencies

```python
Z1_pred,Z2_pred = circuit_3.predict(f1)
```

Dictionary of parameter for EIS and NLEIS can also be obtained using `circuit.extract()`

```python
EIS_dic,NLEIS_dic = circuit_3.extract()
```

`print(circuit)` will show the state information just like impedance.py

```python
print(circuit_3)
```
