```python
def cost_seq(data,model):
    Max = max(abs(data))
    data = data/Max
    model = model/Max
    sum1 = np.sum((data.real-model.real)**2)
    sum2 = np.sum((data.imag-model.imag)**2)
    return(sum1+sum2)
```

# Load data

```python
## fresh Cell
frequencies_30f = np.loadtxt('data/freq_30f.txt')
Z1s_30f = np.loadtxt('data/Z1s_30f.txt').view(complex)
Z2s_30f = np.loadtxt('data/Z2s_30f.txt').view(complex)

## aged cell
frequencies_30a = np.loadtxt('data/freq_30a.txt')
Z1s_30a = np.loadtxt('data/Z1s_30a.txt').view(complex)
Z2s_30a = np.loadtxt('data/Z2s_30a.txt').view(complex)
```

```python
## define the wrapper function to do curve fitting with SSO method and returning the final result
## see the general supplementary notebook for NLEIS toolbox for the implementation of NLEIS toolbox

## fresh cell
def fresh_cell(f1,Z1,f2,Z2,Z2_sd,initial_guess):

    circ_str_1 = 'L0-R0-TDS0-TDP1' ## EIS string
    circ_str_2 = 't(TDSn0,TDPn1)' ## NLEIS string 
    lm = 1e3
    e1 = 2e-10
    e2 = 2e-10
    ub = np.array([1e-6,.1
                   ,1,10,1000,10,100,lm,e2
                   ,.1,.1,10,10,10000,lm,e1
                   ])
    lower_bounds = np.array([0,0,0,0,0,0,0,-lm,-e2,0,0,0,0,0,-lm,-e1])
    upper_bounds = np.ones(16)*ub
    
    bounds = ((lower_bounds), (upper_bounds))
    circuit_30f = EISandNLEIS(circ_str_1,circ_str_2,initial_guess=initial_guess)
    circuit_30f.fit(f1,Z1,Z2_sd, bounds = bounds, opt='max');
    Z1_30f_simul,Z2_30f_simul = circuit_30f.predict(f1)
    return(Z1_30f_simul,Z2_30f_simul,circuit_30f.parameters_)

## aged cell
def aged_cell(f1,Z1,f2,Z2,Z2_sd,initial_guess):

    circ_str_1 = 'L0-R0-TDS0-TDP1' ## EIS string
    circ_str_2 = 't(TDSn0,TDPn1)' ## NLEIS string 
    lm = 1e2
    e1 = 1e-10
    e2 = 0.3
    ub = np.array([1e-6,.1
                   ,1,1,100,10,10000,lm,e2
                   ,.1,.1,10,10,10000,lm,e1
                   ])
    lower_bounds = np.array([0,0,0,0,0,0,0,-lm,-e2,0,0,0,0,0,-lm,-e1])
    upper_bounds = np.ones(16)*ub
    bounds = ((lower_bounds), (upper_bounds))
    circuit_30a = EISandNLEIS(circ_str_1,circ_str_2,initial_guess=initial_guess)
    circuit_30a.fit(f1,Z1,Z2_sd, bounds = bounds, opt='max');
    Z1_30a_simul,Z2_30a_simul = circuit_30a.predict(f1)
    return(Z1_30a_simul,Z2_30a_simul,circuit_30a.parameters_)
```

```python
### Generate the bounds for fresh cell at 30% SOC 
initial_guess = [4.1798e-08,0.0139468,0.0724936,0.000154906,85.6909,0.00141692,17.426,-0.156106,-1e-10,
                0.0448564,0.00143491,0.958662,3.32873e-06,0.505936,-379.2,9.53141e-11]

bounds_fresh = []
for i in range(0, len(initial_guess)):
    if initial_guess[i]>0:
        bounds_fresh.append([initial_guess[i]*0.5, initial_guess[i]*1.5])
    else:
        bounds_fresh.append([initial_guess[i]*1.5, initial_guess[i]*0.5])
```

```python
### Generate the bounds for aged cell at 30% SOC 
initial_guess =[8.73687e-08,0.0136579,0.0197259,0.00738495,5.08418,0.0952935,1877.87,2.3177,0.0635389,0.0113691,
0.00161201,0.576631,0.0318494,614.767,13.0393,3.0205e-13]
bounds_aged = [[initial_guess[i]*0.5, initial_guess[i]*1.5] for i in range(0, len(initial_guess))]
```

```python
# Define the model inputs for sobol sampling 

## aged cell at 30% SOC
problem = {
    'num_vars': 16,
    'names': ['L0','R0','Rpore,c','Rct,c', 'Cdl,c','Ac', 'τc','κc','εc',
              'Rpore,a','Rct,a', 'Cdl,a','Aa', 'τa','κa','εa'],
    'bounds': bounds_aged
}
# Generate samples
param_values = saltelli.sample(problem, 100)
```

## Aged Cell at 30% SoC

```python
f1,Z1,f2,Z2,Z2_sd = data_processing(frequencies_30f,Z1s_30a[1],Z2s_30a[1])

### Making plot to explore the sensitivity of fitting of 30% SOC Aged cell
fig, ax = plt.subplots(2,1,figsize=(5,8))
p_aged = []
cost_aged = []
plot_first(ax[0],Z1,color = 'k')
plot_second(ax[1],Z2,color = 'k')
ax[0].legend(['data'])
ax[1].legend(['data'])
for params in tqdm(param_values):
    Z1_fit,Z2_fit,p0 = aged_cell(f1,Z1,f2,Z2,Z2_sd,params)
    cost_aged.append (cost_seq(Z1,Z1_fit)+cost_seq(Z2,Z2_fit)) 
    p_aged.append(p0)
    plot_first(ax[0],Z1_fit)
    plot_second(ax[1],Z2_fit)
    
plt.tight_layout()
plt.show()
# p will be an array of best fit parameters for each initial guess
```

```python
### Map of cost function
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(cost_aged,'o')
plt.ylabel('Overall Cost')
plt.xlabel('position in the sample')
plt.show()
```

```python
### histogram plot of occurance
fig, ax = plt.subplots(figsize=(4,4))
counts, edges, bars = plt.hist(cost_aged,bins=10)
plt.bar_label(bars)

plt.ylabel('Occurrence Frequency')
plt.xlabel('Overall Cost')
plt.show()
```
